```python
import asyncio
import json
import random
from datetime import datetime, timezone

async def safe_add_episode(i, episode, retries=5):
    for attempt in range(retries):
        try:
            await graphiti.add_episode(
                name=f'Freakonomics Radio {i}',
                episode_body=episode['content'] if isinstance(episode['content'], str) else json.dumps(episode['content']),
                source=episode['type'],
                source_description=episode['description'],
                reference_time=datetime.now(timezone.utc),
            )
            print(f'✅ Added episode: Freakonomics Radio {i} ({episode["type"].value})')
            return
        except Exception as e:
            if "rate limit" in str(e).lower():
                wait_time = (2 ** attempt) + random.random()
                print(f"⚠️ Rate limit hit, retrying in {wait_time:.1f}s...")
                await asyncio.sleep(wait_time)
            else:
                raise

# Run loop safely
for i, episode in enumerate(episodes):
    await safe_add_episode(i, episode)
```

```python
from neo4j import GraphDatabase

# Replace with your actual connection info
NEO4J_URI = "neo4j://127.0.0.1:7687"  # or "neo4j+s://<your-instance>.databases.neo4j.io" if using Aura
NEO4J_USER = "neo4j"
NEO4J_PASSWORD = "password"

# Create driver
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

def get_all_episodes():
    query = """
    MATCH (e:Episode)
    RETURN e.name AS name,
           e.episode_body AS body,
           e.source AS source,
           e.source_description AS description,
           e.reference_time AS reference_time
    ORDER BY e.reference_time ASC
    """
    with driver.session() as session:
        results = session.run(query)
        episodes = [record.data() for record in results]
        return episodes

if __name__ == "__main__":
    episodes = get_all_episodes()
    for ep in episodes:
        print(f"🎙️ {ep['name']} ({ep['source']})")
        print(f"   Description: {ep['description']}")
        print(f"   Body: {ep['body'][:100]}...")  # show first 100 chars
        print(f"   Time: {ep['reference_time']}")
        print("-" * 60)

    driver.close()
```

```python
from neo4j import GraphDatabase

# Replace with your actual connection info
NEO4J_URI = "neo4j://localhost:7687"   # or "neo4j+s://<your-instance>.databases.neo4j.io" if using Aura
NEO4J_USER = "neo4j"
NEO4J_PASSWORD = "password"

driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

def get_all_persons():
    query = """
    MATCH (p:Person)
    RETURN p.name AS name,
           p.position AS position,
           p.state AS state,
           p.previous_role AS previous_role,
           p.previous_location AS previous_location,
           p.term_start AS term_start,
           p.term_end AS term_end
    ORDER BY name ASC
    """
    with driver.session() as session:
        results = session.run(query)
        persons = [record.data() for record in results]
        return persons

if __name__ == "__main__":
    persons = get_all_persons()
    for person in persons:
        print(f"👤 {person['name']}")
        if person.get('position'):
            print(f"   Position: {person['position']}")
        if person.get('state'):
            print(f"   State: {person['state']}")
        if person.get('previous_role'):
            print(f"   Previous Role: {person['previous_role']} in {person.get('previous_location')}")
        if person.get('term_start') or person.get('term_end'):
            print(f"   Term: {person.get('term_start')} → {person.get('term_end')}")
        print("-" * 60)

    driver.close()
```

```python
from neo4j import GraphDatabase
import json
from datetime import datetime, timezone

# Replace with your actual connection info
NEO4J_URI = "neo4j://localhost:7687"   # or "neo4j+s://<your-instance>.databases.neo4j.io"
NEO4J_USER = "neo4j"
NEO4J_PASSWORD = "password"

driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

def add_episode(tx, name, episode_body, source, description, reference_time):
    query = """
    CREATE (e:Episode {
        name: $name,
        episode_body: $episode_body,
        source: $source,
        source_description: $description,
        reference_time: $reference_time
    })
    RETURN e
    """
    tx.run(query, 
           name=name, 
           episode_body=episode_body, 
           source=source, 
           description=description, 
           reference_time=reference_time)

def insert_episodes(episodes):
    with driver.session() as session:
        for i, ep in enumerate(episodes):
            body = ep["content"] if isinstance(ep["content"], str) else json.dumps(ep["content"])
            session.execute_write(
                add_episode,
                f"Freakonomics Radio {i}",
                body,
                ep["type"].name if hasattr(ep["type"], "name") else str(ep["type"]),
                ep["description"],
                datetime.now(timezone.utc).isoformat()
            )
            print(f"✅ Inserted episode: Freakonomics Radio {i}")

if __name__ == "__main__":
    # Example episodes list
    episodes = [
        {
            'content': 'Kamala Harris is the Attorney General of California. She was previously the district attorney for San Francisco.',
            'type': "text",
            'description': 'podcast transcript',
        },
        {
            'content': 'As AG, Harris was in office from January 3, 2011 – January 3, 2017',
            'type': "text",
            'description': 'podcast transcript',
        },
        {
            'content': {
                'name': 'Gavin Newsom',
                'position': 'Governor',
                'state': 'California',
                'previous_role': 'Lieutenant Governor',
                'previous_location': 'San Francisco',
            },
            'type': "json",
            'description': 'podcast metadata',
        },
        {
            'content': {
                'name': 'Gavin Newsom',
                'position': 'Governor',
                'term_start': 'January 7, 2019',
                'term_end': 'Present',
            },
            'type': "json",
            'description': 'podcast metadata',
        },
    ]

    insert_episodes(episodes)
    driver.close()
```

```python
import asyncio
from datetime import datetime, timezone
from graphiti_core import Graphiti
from graphiti_core.nodes import EpisodeType

async def main():
    # Initialize Graphiti
    graphiti = Graphiti(
        uri="neo4j://localhost:7687",
        user="neo4j",
        password="password",
    )

    try:
        # Build necessary indices & constraints (run once)
        await graphiti.build_indices_and_constraints()

        # Example episodes
        episodes = [
            {
                'content': 'Kamala Harris is the Attorney General of California. She was previously the district attorney for San Francisco.',
                'type': EpisodeType.text,
                'description': 'podcast transcript',
            }
        ]

        # Ingest episodes
        for i, ep in enumerate(episodes):
            await graphiti.add_episode(
                name=f"Episode {i}",
                episode_body=ep['content'] if isinstance(ep['content'], str) else json.dumps(ep['content']),
                source=ep['type'],
                source_description=ep['description'],
                reference_time=datetime.now(timezone.utc),
            )
            await asyncio.sleep(20)
            print(f" Added episode: Episode {i}")

        # Perform a hybrid search
        print("\nSearch results for 'Governor of California':")
        results = await graphiti.search("Governor of California", num_results=5)
        for r in results:
            print(f"- {r.fact}")

    finally:
        await graphiti.close()
        print("Connection closed.")

if __name__ == "__main__":
    await main()
```

```python
import json
from graphiti_core.graphiti import EpisodeType
from datetime import datetime, timezone

episodes = [
    {
        'content': 'Kamala Harris is the Attorney General of California. She was previously '
                   'the district attorney for San Francisco.',
        'type': EpisodeType.text,
        'description': 'podcast transcript',
    },
    {
        'content': 'As AG, Harris was in office from January 3, 2011 – January 3, 2017',
        'type': EpisodeType.text,
        'description': 'podcast transcript',
    },
    {
        'content': {
            'name': 'Gavin Newsom',
            'position': 'Governor',
            'state': 'California',
            'previous_role': 'Lieutenant Governor',
            'previous_location': 'San Francisco',
        },
        'type': EpisodeType.json,
        'description': 'podcast metadata',
    },
    {
        'content': {
            'name': 'Gavin Newsom',
            'position': 'Governor',
            'term_start': 'January 7, 2019',
            'term_end': 'Present',
        },
        'type': EpisodeType.json,
        'description': 'podcast metadata',
    },
]

for i, ep in enumerate(episodes, start=1):
    # make sure episode_body is always the right type
    if ep["type"] == EpisodeType.text:
        body = str(ep["content"])  # ensure string
    else:
        body = json.dumps(ep["content"])  # ensure valid JSON string

    await graphiti.add_episode(
        name=f"Episode {i}",
        episode_body=body,
        source=ep["type"],
        source_description=ep["description"],
        reference_time=datetime.now(timezone.utc),
    )
    await asyncio.sleep(20)
    print(f"✅ Added Episode {i}")
```
